```python
# ----------------------------------------------
# PRODIGY INFOTECH DATA SCIENCE INTERNSHIP
# TASK 1: Data Visualization of Population Distribution
# ----------------------------------------------

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
data = pd.read_csv(url)

# Preview data
print("Dataset Preview:")
print(data.head())

```

```python
#  Filter the data for a specific year

year_data = data[data['Year'] == 2020].sort_values(by='Value', ascending=False)

# Take top 10 most populated countries for better visualization
top10 = year_data.head(10)

```

```python
# Bar plot
plt.figure(figsize=(10,6))
sns.barplot(x='Country Name', y='Value', data=top10, palette='viridis')
plt.title('Top 10 Most Populated Countries (2020)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Pie chart
plt.figure(figsize=(8,8))
plt.pie(top10['Value'], labels=top10['Country Name'], autopct='%1.1f%%', startangle=140)
plt.title('Population Share of Top 10 Countries (2020)')
plt.show()

```

```python
# Histogram of population distribution (all countries)
plt.figure(figsize=(8,6))
sns.histplot(year_data['Value'], bins=20, kde=True)
plt.title('Global Population Distribution (2020)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
```

```python
# Step 4: Insights
# ----------------------------------------------
print("\nInsights:")
print("1️⃣ China and India have the highest population globally.")
print("2️⃣ Population distribution is highly skewed — a few countries have very large populations.")
print("3️⃣ The majority of countries have populations below 100 million.")

```
